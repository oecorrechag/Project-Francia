```python
from fuzzywuzzy import fuzz
import pandas as pd
import numpy as np

pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
dataframecolumn = pd.DataFrame(["apple","tb",'Bogota','Medellin','Purisima','Purificacion','Uramita','Urumita'])
dataframecolumn.columns = ['Match']
dataframecolumn['Key'] = 1
dataframecolumn
```

```python
compare = pd.DataFrame(["adfad","apple","asple","tab", 'Bogota', 'bogota', 'Bogota DC', 'bogota dc', 'medellin', 'Medellin', 'Medellín', 'purisiima','prurificacion','urumita'])
compare.columns = ['compare']
compare['Key'] = 1
compare
```

```python
x = np.arange(len(compare))
```

```python
compare['compare2'] = compare['compare'].str.capitalize()
compare['id'] = x
compare
```

```python
combined_dataframe = dataframecolumn.merge(compare,on="Key",how="left")
# combined_dataframe = combined_dataframe[~(combined_dataframe.Match==combined_dataframe.compare)]
combined_dataframe
```

```python
def partial_match(x,y):
    return(fuzz.ratio(x,y))
partial_match_vector = np.vectorize(partial_match)
```

```python
combined_dataframe['score'] = partial_match_vector(combined_dataframe['Match'], combined_dataframe['compare2'])
combined_dataframe
```

```python
combined_dataframe = combined_dataframe[combined_dataframe.score >= 80]
combined_dataframe
```
